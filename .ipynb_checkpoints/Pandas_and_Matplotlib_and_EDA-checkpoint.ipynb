```python
#constructing a dataframe from scratch
import string
spam_corpus = map(string.split, ["hello world", "hello python","both python of this world are magestic"])
print 'spam_corups=',spam_corpus

words = [word for doc in spam_corpus for word in doc]
print 'words=', words

unique_words = set([word for doc in spam_corpus for word in doc])
print 'unique_words=', unique_words

word_counts = [(word, map(lambda doc: doc.count(word), spam_corpus)) for word in unique_words]
print 'word_counts=', word_counts

spam_bag_of_words = pd.DataFrame(dict(word_counts))

print 'spam_bag_of_words=\n', spam_bag_of_words
```

```python
#combining datasets
#concatenation
other_corpus = map(string.split, [ "python was named after monty python", "Guido van Rossum was creater of python" ])

unique_words = set([ word for doc in other_corpus for word in doc ])

word_counts = [ (word, map(lambda doc: doc.count(word), other_corpus)) for word in unique_words ]

other_bag_of_words = pd.DataFrame(dict(word_counts))

pd.concat([spam_bag_of_words, other_bag_of_words], ignore_index=True)
```

```python
#join
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
print left

print right

print 'lef outer join'
print pd.merge(left, right, how='left', on=['key1', 'key2'])

print 'right outer join'
print pd.merge(left, right, how='right', on=['key1', 'key2'])

print 'outer join'
print pd.merge(left, right, how='outer', on=['key1', 'key2'])

print 'inner join'
print pd.merge(left, right, how='inner', on=['key1', 'key2'])
```
